```python
#XGBoost Algorithm
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
#ONE HOT ENCODING for Nominal Data
dataset=pd.get_dummies(dataset, drop_first=True)
dataset=dataset.astype(int)
dataset
```

```python
#SPLITTING INDEPENDENT (INPUT) AND DEPENDENT (OUTPUT) VARIABLE
#To view columns in the dataset
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
import xgboost as xgb
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'loss':['squared_error', 'absolute_error', 'huber', 'quantile'],'num_leaves': [31],'criterion':['friedman_mse'], 'max_features':[None], 'n_estimators': [10, 25, 50,100, 500], 'random_state':[None]}

grid = GridSearchCV(xgb.XGBRegressor(), param_grid, refit=True, verbose=3, n_jobs=-1)

grid.fit(independent, dependent)
```

```python
# print best parameter after tuning
#print(grid.best_params_)
re=grid.cv_results_
print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
```

```python
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("Future_Prediction={}".format(Future_Prediction))
```

```python
# Saving the model
import pickle
filename= "finalized_model_XGBoost_Grid_insurance_charges.sav"
```

```python
pickle.dump(grid,open(filename,'wb'))
load_model=pickle.load(open("finalized_model_XGBoost_Grid_insurance_charges.sav",'rb'))
```
