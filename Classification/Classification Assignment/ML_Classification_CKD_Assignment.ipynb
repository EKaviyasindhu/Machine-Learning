```python
#importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Reading the dataset
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset, drop_first=True)
dataset=dataset.astype(int)
dataset
```

```python
dataset.columns
```

```python
indep = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv','wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 
                 'rbc_normal', 'pc_normal','pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes','appet_yes', 'pe_yes', 'ane_yes']]
dep = dataset['classification_yes']
```

```python
indep
```

```python
dep
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
#StandardScaler - Preprocessing
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# 1.SUPPORT VECTOR MACHINE - STARTS
```

```python
from sklearn.svm import SVC

#GridSearchCV
from sklearn.model_selection import GridSearchCV
param_grid = {'kernel':['linear','rbf','poly','sigmoid'],'gamma':['auto','scale'],'C':[10,100,1000,2000,3000]}
grid_SVM = GridSearchCV(SVC(probability=True), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
   
# fitting the model for grid search 
grid_SVM.fit(X_train, y_train)

# print best parameter after tuning 
print(grid_SVM.best_params_)
re=grid_SVM.cv_results_
grid_predictions = grid_SVM.predict(X_test)

#print confusion_matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions)

#Print classification report 
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)

#Print f1_score
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_predictions,average='weighted')

#roc_auc_score
from sklearn.metrics import roc_auc_score
roc_auc_score=roc_auc_score(y_test,grid_SVM.predict_proba(X_test)[:,1])

table=pd.DataFrame.from_dict(re)

#PRINT OUTPUT VALUES OF SUPPORT VECTOR MACHINE 
print("\n PRINT OUTPUT VALUES OF SUPPORT VECTOR MACHINE ")
print("\n The confusion Matrix:\n",cm)
print("\n The Classification report:\n\n",clf_report)
print("\n The f1_macro value for best parameter {}:".format(grid_SVM.best_params_),f1_macro)
print("\n The ROC_AUC_SCORE:",roc_auc_score)
table
```

```python
# 1.SUPPORT VECTOR MACHINE - ENDS
```

```python
# III.CategoricalNB
```

```python
#Check if y_train Has Multiple Classes - Categorical Naive Bayes requires at least two unique classes in y_train. If y_train has only one class, the model cannot make predictions.
print(np.unique(y_train)) 

#Ensure X_train and X_test Have the Right Shape - Expected shape: (samples, features), e.g., (100, 5).
print(X_train.shape, X_test.shape)

#Ensure Your Features Are Categorical (Integers) - CategoricalNB only works with categorical (discrete) features, but X_train may contain continuous (float) values.
#If X_train contains continuous values, discretize them
print(pd.DataFrame(X_train).dtypes)

#Check for NaNs in X_train and X_test
print(np.isnan(X_train).sum(), np.isnan(X_test).sum())
```

```python
# Apply KBinsDiscretizer to convert continuous values to discrete bins
from sklearn.preprocessing import KBinsDiscretizer

discretizer = KBinsDiscretizer(n_bins=10, encode='ordinal', strategy='uniform')
X_train = discretizer.fit_transform(X_train)
X_test = discretizer.transform(X_test)
```

```python
# Ensure no NaNs are introduced
X_train = np.nan_to_num(X_train)
X_test = np.nan_to_num(X_test)
```

```python
# III.CategoricalNB
```

```python
from sklearn.naive_bayes import CategoricalNB

#GridSearchCV
from sklearn.model_selection import GridSearchCV
#param_grid = {'alpha': [0.01, 0.1, 0.5, 1.0, 5.0, 10.0],'fit_prior': [True, False]}
param_grid = {'alpha': [0.01, 0.1, 0.5, 1.0, 5.0, 10.0]}
grid = GridSearchCV(CategoricalNB(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 

# fitting the model for grid search 
grid.fit(X_train, y_train)

# print best parameter after tuning 
print(grid.best_params_)
re=grid.cv_results_
grid_predictions = grid.predict(X_test)

#print confusion_matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions)

#Print classification report 
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)

#Print f1_score
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_predictions,average='weighted')

#roc_auc_score
from sklearn.metrics import roc_auc_score
roc_auc_score=roc_auc_score(y_test,grid.predict_proba(X_test)[:,1])

table=pd.DataFrame.from_dict(re)

#PRINT OUTPUT VALUES OF NAIVE BAYES - CategoricalNB
print("\n PRINT OUTPUT VALUES OF NAIVE BAYES - CategoricalNB")
print("\n The confusion Matrix:\n",cm)
print("\n The Classification report:\n\n",clf_report)
print("\n The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
print("\n The ROC_AUC_SCORE:",roc_auc_score)
table
```
