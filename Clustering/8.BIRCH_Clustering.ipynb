```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```python
X = dataset.iloc[:,3:5].values
X
```

```python
#Apply BIRCH Clustering
from sklearn.cluster import Birch
BRC = Birch(n_clusters=3)
y_BRC = BRC.fit_predict(X)
```

```python
y_BRC
```

```python
#Calculate Silhouette Score
from sklearn.metrics import silhouette_score
sil_score = silhouette_score(X, y_BRC)
print(f"Silhouette Score: {sil_score:.3f}")
```

```python
#Append 'Cluster_group' Column to the Dataset
supervised = dataset
supervised['Cluster_group']=y_BRC
supervised
```

```python
#Save the Dataset into a csv file
supervised.to_csv("cluster_BRC.csv",index=False)
```

```python
#TO VISUALIZE THE CLUSTERED DATA
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], 
                   fit_reg=False, legend=True, legend_out=True)
```
