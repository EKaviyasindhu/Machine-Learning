```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```python
X = dataset.iloc[:,3:5].values
X
```

```python
#Append 'Cluster_group' Column to the Dataset
```

```python
from sklearn.cluster import MeanShift
MS = MeanShift(bandwidth=2).fit(X)
y_MS = MS.fit_predict(X)
```

```python
y_MS
```

```python
supervised = dataset
supervised
```

```python
supervised['Cluster_group']=y_MS
supervised
```

```python
#Save the Dataset into a csv file
```

```python
supervised.to_csv("cluster_MS.csv",index=False)
```
