```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```python
X = dataset.iloc[:,3:5].values
X
```

```python
#Append 'Cluster_group' Column to the Dataset
```

```python
#Estimate Bandwidth
from sklearn.cluster import estimate_bandwidth
bandwidth = estimate_bandwidth(X, quantile=0.2) 
```

```python
#Apply MeanShift Clustering
from sklearn.cluster import MeanShift
MS = MeanShift(bandwidth=bandwidth)
y_MS = MS.fit_predict(X)
```

```python
y_MS
```

```python
#Calculate Silhouette Score
from sklearn.metrics import silhouette_score

n_clusters = len(set(y_MS))
print(f"Number of clusters: {n_clusters}")

if n_clusters > 1:
    sil_score = silhouette_score(X, y_MS)
    print(f"Silhouette Score: {sil_score:.3f}")
else:
    print("Only one cluster found — Silhouette Score not applicable.")
```

```python
supervised = dataset
supervised
```

```python
supervised['Cluster_group']=y_MS
supervised
```

```python
#Save the Dataset into a csv file
```

```python
supervised.to_csv("cluster_MS.csv",index=False)
```
