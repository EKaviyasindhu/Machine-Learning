```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```python
X = dataset.iloc[:,3:5].values
X
```

```python
#Apply OPTICS Clustering
from sklearn.cluster import OPTICS
OP = OPTICS(min_samples=5, xi=0.05, min_cluster_size=0.1)
y_OP = OP.fit_predict(X)
```

```python
y_OP
```

```python
#Calculate Silhouette Score
from sklearn.metrics import silhouette_score
sil_score = silhouette_score(X, y_OP)
print(f"Silhouette Score: {sil_score:.3f}")
```

```python
#Append 'Cluster_group' Column to the Dataset
supervised = dataset
supervised['Cluster_group']=y_OP
supervised
```

```python
#Save the Dataset into a csv file
supervised.to_csv("cluster_OP.csv",index=False)
```
